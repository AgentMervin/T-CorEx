# create random covariance matrix

```python
def generate_covariance(n, nb=1):
    # TODO: number of blocks
    A = np.zeros((n, n))
    for i in range(n):
        A[i, :i+1] = np.random.uniform(-1, 1, size=(i+1,))
        A[i][i] = np.random.uniform(0.1, 1)
    return np.dot(A, A.T)
```

```python
nt = 100
nv = 100
myu = np.zeros((nv,))
sigma1 = generate_covariance(nv)
sigma2 = generate_covariance(nv)

X = []
for i in range(nt):
    if i < nt // 2:
        sigma = sigma1
    else:
        sigma = sigma2    
    ns = np.random.randint(4, 10)
    X.append(np.random.multivariate_normal(myu, sigma, size=(ns,)))
```

```python
corex = time_corex.Corex(n_hidden=10,
                         max_iter=300,
                         optimizer=Adam(lr=0.001),
                         regularizer=L2Reg(10))
corex.fit(X)
```

```python
print sum(corex.tc)
```

## train single corexes

```python
Cs = []
tcs = []
for t in range(nt):
    print "Training corex {} / {}".format(t+1, nt)
    c = linearcorex.Corex(n_hidden=10,
                          max_iter=300)
    c.fit(X[t])
    Cs.append(c)
    tcs.append(c.tc)
```
